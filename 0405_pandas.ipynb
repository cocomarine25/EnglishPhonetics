### Importing libraries

```python
import pandas as pd
```

### Loading csv file

```python
df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")   # csv는 xlsx보다 심플한 텍스트 파일(용량이 더 작음)
df
# csv : comma seperated value
```

```python
print(df.size); print(df.ndim); print(df.shape)
# df.ndim : 엑셀, csv는 무조건 2차원이다. 3차원일 수 없다.
# df.shape : 행이 무조건 먼저. 행 : 17000, 열 : 9
```

```python
df.describe()

# standard deviation(std)이 1이면 무슨 뜻이고 12면 무슨 뜻일까? 정규분포값을 생각하면 됨. 평균 아이큐가 100, std가 10이라면 90 ~ 110
```

### Constructing a DataFrame using dict

```python
# Some input data - a small sample of the iris dataset
# 데이터를 직접 손으로 작성하는 방법
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df
```

### Constructing a DataFrame using pd.Series

```python
# 리스트를 사용해서 손으로 직접 데이터를 작성하는 방법
# 리스트를 pd.Series라고 이름 붙여서 넣음
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

### Changing column names

```python
df.columns
```

```python
# Change the columns
# 대문자였던 SPECIES가 소문자로 바뀜
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

### Accessing columns

```python
# Access one column
df.loc[:, 'sepal_length']
```

```python
# Access one column (short-hand)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row
df.loc[3, :]    # , :은 3번째 로우 전체를 전부 다 가져오기
```

```python
# Access two rows
df.loc[[3, 0], :]   # 4번째와 0번째 로우를 모두 가져오기
```

```python
# Access a range of rows
df.loc[2:5, :]    # 2, 3, 4의 로우와 모든 로우 가져오기
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4]
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1    # 새로운 컬럼을 만들고 모두 1로 써라
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]    # 새로운 컬럼을 만들고 1,2,3,4,5를 넣어주기
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column   # 칼럼 없애기
df
```

### Adding/Removing new rows

```python
df.loc[5, :] = ''   # 5번째 로우의 모든 칼럼을 없애라
df
```

```python
df = df.drop(5, axis=0) # axis=0: row
df
```
